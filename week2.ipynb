```python
# 11. Sort the data w.r.t price_find the details of the most and the least expensive cars

def find_min_max(df,column):
    min_val=df[column].min()
    max_val=df[column].max()
    
    min_row=df[df[column]==min_val]
    max_row=df[df[column]==max_val]
    
    return min_val,max_val,min_row,max_row

min_val,max_val,min_row,max_row=find_min_max(df,'Price in thousands')
print("most expensive car ",max_row)
print("\n\n\n")
print("least expensive car ",min_row)
```

```python
# 12. Write a function to find min and max values of any column

def find_min_max(df,colum_name):
    min_val=df[colum_name].min()
    max_val=df[colum_name].max()
    return min_val,max_val


min_price,max_price=find_min_max(df,"Price in thousands")
min_sales,max_sales=find_min_max(df,'Sales in thousands')

print(f"min price: {min_price},max price: {max_price}")
print(f"min sales: {min_sales},max sales: {max_sales}")
    
```

```python
# 26.Split the data set into train and test set 70% train set 10% test set

import pandas as pd
from sklearn.model_selection import train_test_split

df=pd.read_csv("Car_sales.csv")

x=df.drop(columns=["Price in thousands"])
y=df["Price in thousands"]
     
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.3,random_state=42)

x_test,x_val,y_test,y_val=train_test_split(x_temp,y_temp,test_size=2/3,random_state=42)

train_percentage=((len(x_train))/len(df))*100
test_percentage=((len(x_test))/len(df))*100
validation_percentage=((len(x_val))/len(df))*100


print("shape of the train data :",x_train.shape,y_train.shape)
print("shape of the test data :",x_test.shape,y_test.shape)
print("shape of the validation data :",x_val.shape,y_val.shape)
print("\n\n")


print("percentage of train data :",train_percentage,"%")
print("percentage of test data :",test_percentage,"%")
print("percentage of validation :",validation_percentage,"%")
```
