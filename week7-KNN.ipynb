```python

from sklearn.model_selection import train_test_split
y=df_sample["label"]
x=df_sample.drop(columns=['label'])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=45)

train_per=(len(x_train))/(len(df_sample))*100
test_per=(len(x_test))/len(df_sample)*100
print(train_per)
print(test_per)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=5,p=2,metric='minkowski')

y_pred=knn.fit(x_train,y_train)
```

```python
y_test_pred=knn.predict(x_test)
print(y_test_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

ac=accuracy_score(y_test,y_test_pred)
cm=confusion_matrix(y_test,y_test_pred)
cr=classification_report(y_test,y_test_pred)
print(ac)
print(f'accuracy : {ac *100: .2f}')

print(f"\n confused matrix\n : {cm}")
print(f"\n classification report \n:{cr}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

#  Predict test set
y_pred = knn.predict(x_test)

#  Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges')
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()

#  Classification Report
print("Classification Report:\n")
print(classification_report(y_test, y_pred))
```
