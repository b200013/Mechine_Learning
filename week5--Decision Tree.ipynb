## Handling Missing Values

```python
#Replace the missing values in Embarked column by 'S'


df["Embarked"].fillna("S", inplace=True)

df.isnull().sum()
```

```python
#Impute the missing values in Fare column with median value
# df.isnull().sum()

# df["Fare"].isnull()

df["Fare"].fillna(df["Fare"].median(), inplace=True)

# df.isnull().sum()
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
import numpy as np
mean=df["Age"].mean()

std=df["Age"].std()
age_null_count=df['Age'].isnull().sum()

random_ages=np.random.normal(loc=mean,scale=std,size=age_null_count,)

random_ages=np.abs(random_ages)

df.loc[df['Age'].isnull(),'Age']=random_ages

age_null_count,df.isnull().sum()
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1


df["FamilySize"]=df["Parch"]+df["SibSp"]+1
df
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
import numpy as np
# df["IsAlone"]=df["FamilySize"] == 1  (or)

df["IsAlone"]=np.where(df["FamilySize"] == 1,True,False )
df
```

```python


# df["Has_Cabin"]=df["Cabin"].notnull()

df["Has_cabin"]=np.where(df["Cabin"].isnull(),False,True)
df
```

## Create a new attribute Has_cabin using the attribue Cabin. 
#if Cabin == NaN the Has_cabin = False else Has_cabin = True


```python

#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin





df.drop(columns=['PassengerId',"Name","Ticket","Cabin"],inplace=True)
df
```

## Model Building

```python
#Prepare X and y

# y=df["Survived"]
# x=df[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked","FamilySize","IsAlone","Has_Cabin"]]

y=df["Survived"]
x=df.drop(columns=["Survived"])
```

```python
# Label encoding of categorical features
import pandas as pd
from sklearn.preprocessing import LabelEncoder

le_sex=LabelEncoder()
le_Em=LabelEncoder()


df["Sex"]=le_sex.fit_transform(df["Sex"])


df
```

```python
# one hot encoding 

df=pd.get_dummies(df,columns=["Embarked"],drop_first=True )

df
```

```python
df["IsAlone"] = df["IsAlone"].astype(int)
df["Has_cabin"] = df["Has_cabin"].astype(int)
df["Embarked_Q"]=df["Embarked_Q"].astype(int)
df["Embarked_S"]=df["Embarked_S"].astype(int)
df
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split


y=df["Survived"]
x=df.drop(columns=["Survived"])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)

train_per=(len(x_train)/(len(df)))*100
test_per=(len(x_test)/len(df))*100

print(train_per)
print(test_per)

```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library


from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)

train_accuracy = model.score(x_train, y_train)
print(f"Training Accuracy: {train_accuracy:.2f}")
```

```python
#Print the tree
# Importing required packages for visualization
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(model,feature_names=df.columns,class_names=["Survived","Not Survived"] ,filled=True)
plt.show()
```

```python
# plotting tree with max_depth=3

plt.figure(figsize=(20,10))
plot_tree(model,feature_names=df.columns,class_names=['Not Surivived','Survived'],filled=True,max_depth=3)
plt.show()
```

```python
# Check the evaluation metrics of our default model on validation set - Confusion Matrix, Accuracy, Precision, Recall, F1 score

# Importing classification report and confusion matrix from sklearn metrics

# Making predictions

# Printing classification report


from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,precision_score,recall_score,f1_score

y_pred=model.predict(x_test)
print("confusion matrix\n")

print(confusion_matrix(y_test,y_pred))
print("\n\n\n\n")

print("accuracy score :",accuracy_score(y_test,y_pred))
print("precision :",precision_score(y_test,y_pred))
print("recall :",recall_score(y_test,y_pred))
print('f1 :',f1_score(y_test,y_pred))

print("\n\nclassififcation report :\n \n",classification_report(y_test,y_pred,target_names=['Not Survived', 'Survived']))






```
