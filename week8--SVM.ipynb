```python
import matplotlib.pyplot as plt
import seaborn as sns
c=df.drop(columns=['Species'])
plt.figure(figsize=(12, 8))
for i,col in enumerate(c):
  
    plt.subplot(2, 2, i+1)
    sns.histplot(df[col],kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.suptitle('Histograms of All Input Features', fontsize=16, y=1.02)
plt.show()
```

```python

import seaborn as sns


# Create a FacetGrid for scatter plot
g = sns.FacetGrid(df, hue="Species", height=5)
g.map(sns.scatterplot, "SepalLengthCm", "SepalWidthCm")
g.add_legend()

# Add titles and labels
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Sepal Length vs Sepal Width by Species")
plt.show()
```

```python


x=df.drop(columns=["Species"])
y=df['Species']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)

tr_per=(len(x_train))/(len(df))*100
ts_per=(len(x_test))/(len(df))*100

print(tr_per)
print(ts_per)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

model = SVC(C=1)  # or 'rbf', 'poly', etc.
model.fit(x_train, y_train)

y_test_pred = model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_test_pred))
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score

print("confusion matrix :\n",confusion_matrix(y_test,y_test_pred))

print("precision \n",precision_score(y_test,y_test_pred,average='macro'))
print("recall :",recall_score(y_test,y_test_pred ,average='macro'))
print('f1 score :',f1_score(y_test,y_test_pred,average='macro'))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_test_pred))
```
